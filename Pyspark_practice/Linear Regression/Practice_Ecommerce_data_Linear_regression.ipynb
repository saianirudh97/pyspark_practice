```
from pyspark.sql import SparkSession
from pyspark.ml.regression import LinearRegression
```

```
spark = SparkSession.builder.appName('LinearRegression_Ecommerce').getOrCreate() 
```

```
data = spark.read.csv('/FileStore/tables/4_1_Ecommerce_Customers.csv',inferSchema=True,header=True)
```

```
data.show()
```

```
data.printSchema()
```

```
data.describe().show()
```

```
data.columns
```

```
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
```

```
assembler =VectorAssembler(inputCols=[
 'Avg Session Length',
 'Time on App',
 'Time on Website',
 'Length of Membership',
 'Yearly Amount Spent'],outputCol='features')
```

```
output = assembler.transform(data)
```

```
output.show()
```

```
final_data = output.select('features'
                           ,'Yearly Amount Spent')
```

```
final_data.show()
```

```
train_data,test_data = final_data.randomSplit([0.7,0.3])
```

```
train_data.describe().show()
```

```
test_data.describe().show()
```

```
lr = LinearRegression(featuresCol='features',labelCol='Yearly Amount Spent')
lr_model = lr.fit(train_data)
```

```
test_results = lr_model.evaluate(test_data)
```

```
test_results.residuals.show()
```

```
test_results.rootMeanSquaredError
```

```
test_results.r2
```

Predicting the yearly amount spent

```
unlabelled_data = test_data.select('features')
```

```
unlabelled_data.show()
```

```
predictions = lr_model.transform(unlabelled_data)
```
